```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
path = "shampoo.csv"
data = pd.read_csv(path,parse_dates=['Month'])
print(data.head(10))
```

```python
plt.figure(figsize=(17,8))
plt.plot(data.Month,data.Sales)
plt.title("Data of the monthly shampoo sales")
plt.xticks(rotation = 45)
plt.show()
```

```python
#Yes, the graph has a very highly defined positive trend

#The graph shows some seasonality patterns as we can see that roughly every october/november, 
#the graph peaks, hence we can say the graph has a yearly seasonality

#There are no outliers in the graph
```

```python
from datetime import datetime
BASE_DATE = data.Month[0]
def getDateAsMonthCount(date1,date2):
    temp = date1.split('-')
    year1 = int(temp[0])
    month1 = int(temp[1])
    temp2 = date2.split('-')
    year2 = int(temp2[0])
    month2 = int(temp2[1])
    return (year2-year1)*12 + month2-month1
def getMonthCountAsDate(date1,count2):
    temp = date1.split('-')
    year1 = int(temp[0])
    month1 = int(temp[1])    
    years = int(count2/12)
    months = int(count2%12)
    year2 = year1+years
    month2 = months+month1
    temp = str(year2)+"-"+"{0}".format(str(month2).zfill(2))
    return str(temp)
```

```python
y=np.array(data['Sales'].dropna().values, dtype=float)
X= np.array([getDateAsMonthCount(BASE_DATE,x) for x in data.Month]).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(X,y)
plt.figure(figsize=(17,8))
plt.plot(data.Month,data.Sales)
plt.plot(lr.coef_*X+lr.intercept_)
plt.title("Data of the monthly shampoo sales")
plt.xticks(rotation = 45)
plt.show()
```

```python
#Taking 60-70 point
X_test= np.arange(60,80,1).reshape(-1,1)
X_test_date = [getMonthCountAsDate(BASE_DATE,x) for x in X_test]
Y_pred = lr.predict(X_test)
```

```python
plt.figure(figsize=(17,8))
plt.plot(X_test,Y_pred)
plt.xticks(X_test.ravel(),X_test_date)
plt.show()
```

```python
import random
X_random = np.array(random.sample(range(0, 100), 10)).reshape(-1,1)
X_test_date = [getMonthCountAsDate(BASE_DATE,x) for x in X_random]
Y_random_pred = lr.predict(X_random)
for x,y in zip(X_test_date,Y_random_pred):
    print("Date :" ,x, "  Monthly shampoo sales: ",int(y))
```
